```python
def stock_talk(RIC, TGT = 'none'):
    market_data_fields = [ek.TR_Field('CF_CLOSE'), ek.TR_Field('TR.Price52WeekHigh'), ek.TR_Field('TR.Price52WeekLow'),
                   ek.TR_Field('TR.CompanyMarketCap', params = {'Scale':6}), ek.TR_Field('TR.EV', params = {'Scale':6})]
    market_data = ek.get_data(RIC, fields=market_data_fields)[0]
    market_data = market_data.rename(columns = {"Instrument": 'Symbol', 'CF_CLOSE': 'Last Close', "Company Market Cap": 'Market Cap (M)', 
                                                'Enterprise Value (Daily Time Series)':'Enterprise Value (M)' })
    market_data = market_data.set_index('Symbol')
    market_data['Market Cap (M)'] = '${:,}'.format(np.round(market_data['Market Cap (M)'].iloc[0], 2))
    market_data['Enterprise Value (M)'] = '${:,}'.format(np.round(market_data['Enterprise Value (M)'].iloc[0], 2))
    
    box_1 = market_data.T
    
    ws_recommendation_fields = [ek.TR_Field('TR.NumOfStrongBuy'), ek.TR_Field('TR.NumOfBuy'), ek.TR_Field('TR.NumOfHold'),
                             ek.TR_Field('TR.NumOfSell'), ek.TR_Field('TR.NumOfStrongSell')]
    ws_recommendations= ek.get_data(RIC, fields=ws_recommendation_fields)[0]
    ws_recommendations = ws_recommendations.rename(columns = {'Recommendation - Number Of Strong Buy':'Strong Buy', 
                                                             'Recommendation - Number Of Buy': 'Buy', 'Recommendation - Number Of Hold': 'Hold',
                                                             'Recommendation - Number Of Sell':'Sell', 'Recommendation - Number Of Strong Sell':'Sell'})
    
    ws_recommendations = ws_recommendations.set_index('Instrument')
    
    
    box_2 = ws_recommendations.T
    
    tgt_price_data = [ek.TR_Field('TR.PriceTargetMean'), ek.TR_Field('TR.PriceTargetNumIncEstimates'), ek.TR_Field('TR.PriceTargetHigh'),
                     ek.TR_Field('TR.PriceTargetLow'), ek.TR_Field('TR.PriceTargetStdDev'), ek.TR_Field('TR.LTGMean')]
    
    tgt_price = ek.get_data(RIC, fields= tgt_price_data)[0]
    if isinstance(TGT, float)==True:
        tgt_price['Lido Estimate'] = TGT
        tgt_price['Upside'] = round(((tgt_price['Lido Estimate'].iloc[0]/ market_data['Last Close'].iloc[0])-1)*100,2)
    else:
        tgt_price['Lido Estimate'] = 'N/A'
        tgt_price['Upside'] = round((( tgt_price['Price Target - Mean'].iloc[0]/ market_data['Last Close'].iloc[0])-1)*100,2)
    
    tgt_price['Price Target - Standard Deviation'] = round(float(tgt_price['Price Target - Standard Deviation'].iloc[0]), 2)
    tgt_price['Long Term Growth - Mean'] = round(float(tgt_price['Long Term Growth - Mean'].iloc[0]), 2)
    tgt_price['Price Target - Mean'] = round(float(tgt_price['Price Target - Mean'].iloc[0]),2) 
        
    tgt_price = tgt_price.set_index('Instrument')
    
    tgt_price = tgt_price.T
    
    
    mults_data_ltm = [ek.TR_Field('TR.EVToSales'), ek.TR_Field('TR.EVToEBITDA'), ek.TR_Field('TR.EVToEBIT'),
                 ek.TR_Field('TR.PE'), ek.TR_Field('TR.PriceToBVPerShare'), ek.TR_Field('TR.ROAPercentTrailing12M'),
                     ek.TR_Field('TR.ReturnonAvgTotEqtyPctNetIncomeBeforeExtraItemsTTM'), ek.TR_Field('TR.HistPEG')]
    
    mults_data_ntm = [ek.TR_Field('TR.FwdEVToSales'), ek.TR_Field('TR.FwdEVToEBITDA'), 
                      ek.TR_Field('TR.FwdEVToEBIT'), ek.TR_Field('TR.FwdPE'), 
                      ek.TR_Field('TR.FwdPriceToBVPerShare'), ek.TR_Field('TR.ROAMean'),
                     ek.TR_Field('TR.ROEMean'), ek.TR_Field("TR.FwdPEG")]
    
    
    mults_ltm = ek.get_data(RIC, mults_data_ltm)[0]
    mults_ntm = ek.get_data(RIC, mults_data_ntm)[0]
    
    
    mults_ltm = mults_ltm.replace(pd.NA, 'NaN')
    mults_ntm = mults_ltm.replace(pd.NA, 'NaN')
    mults_ltm_vals = list(mults_ltm.iloc[0])
    mults_ntm_vals = list(mults_ntm.iloc[0])

    fields = ['EV/SALES', 'EV/EBITDA', 'EV/EBIT', 'P/E', 'P/B', 'ROA (%)', 'ROE (%)', 'PEG']
    mults_ltm_vals = [np.round(i,2) if isinstance(i, numbers.Number)==True else i for i in mults_ltm_vals[1:]]
    mults_ntm_vals = [np.round(i,2) if isinstance(i, numbers.Number)==True else i for i in mults_ntm_vals[1:]]
    
    box_3_prep = pd.DataFrame({'Multiples': fields, 'LTM':mults_ltm_vals, 'NTM':mults_ntm_vals})

    fig = make_subplots(
    rows=2, cols=3,
    vertical_spacing=0.005,
    specs=[[{'type': 'table'}, {'type': 'table', 'rowspan':2}, {'type': 'table', 'rowspan':2}],
           [{'type': 'table'}, None, None]])
    fig.add_trace(go.Table(
        columnorder = [1,2],
        columnwidth = [70,80],
        header = dict(values = ['', '<b>Market Data</b>'], 
                      fill_color = '#003C5A',
                      line_color = 'Black',
                      height = 30 ,
                      font = dict(size =  15, color = 'white', family = 'NanumGothic')),
                cells = dict(values= [box_1.index, box_1['{}'.format(RIC)]],
                     fill_color = ['#e5e5e5', 'white'],
                     line_color = 'Black',
                     font = dict(size = 15, color = ['#003C5A', 'Black'], family = 'NanumGothic'),
                     height = 30,
                     align = ['left','center'])), row = 1, col = 1)
    
    fig.add_trace(go.Table(
        columnorder = [1,2],
        columnwidth = [70,60],
        header = dict(values = ['', '<b>Wall St. Recos</b>'], 
                      fill_color = '#003C5A',
                      line_color = 'Black',
                      height = 30 ,
                      font = dict(size =  15, color = 'white', family = 'NanumGothic')),
                cells = dict(values= [box_2.index, box_2['{}'.format(RIC)]],
                     fill_color = ['#e5e5e5', 'white'],
                     line_color = 'Black',
                     font = dict(size = 15, color = ['#003C5A', 'Black'], family = 'NanumGothic'),
                     height = 30,
                     align = ['left','center'])), row = 2, col = 1)
    
    

    fig.add_trace(go.Table(
        columnorder = [1,2],
        columnwidth = [70,80],
        header = dict(values = ['', '<b>Price Targets</b>'], 
                      fill_color = '#003C5A',
                      line_color = 'Black',
                      height = 30 ,
                      font = dict(size =  15, color = 'white', family = 'NanumGothic')),
                cells = dict(values= [tgt_price.index, tgt_price['{}'.format(RIC)]],
                     fill_color = ['#e5e5e5', 'white'],
                     line_color = 'Black',
                     font = dict(size = 15, color = ['#003C5A', 'Black'], family = 'NanumGothic'),
                     height = 30,
                     align = ['left','center'])), row = 1, col = 2)
    
    
    fig.add_trace(go.Table(
        columnorder = [1,2,3],
        columnwidth = [70,40,40],
        header = dict(values = ['<b>Multiples</b>', '<b>LTM</b>', '<b>NTM</b>'], 
                      fill_color = '#003C5A',
                      line_color = 'Black',
                      height = 30 ,
                      font = dict(size =  15, color = 'white', family = 'NanumGothic')),
                cells = dict(values= [box_3_prep['Multiples'], box_3_prep['LTM'],box_3_prep['NTM'] ],
                     fill_color = ['#e5e5e5', 'white', 'white'],
                     line_color = 'Black',
                     font = dict(size = 15, color = ['#003C5A', 'Black', 'Black'], family = 'NanumGothic'),
                     height = 30,
                     align = ['left','center'])), row = 1, col = 3) 
    
    
    name = ek.get_data(RIC, fields = 'TR.CommonName')[0]
    
    fig.update_layout(
        margin=dict(l=48, r=48, t=100, b=10),
        paper_bgcolor='white',
        plot_bgcolor='white',
        width = 1080,
        height = 768,
        legend_orientation="h",
        title= dict(text="{} ({})".format(name['Company Common Name'].iloc[0], RIC), font = dict(family = 'NanumGothic', size = 20),
                   x = 0.5, y = 0.93))
        
    fig.layout.images = [dict(
        source="https://jgarden79.github.io/PM_vis/NewLidoLogo@300x.png",
        xref="paper", yref="paper",
        x=0.12, y=1.14,
        sizex=0.3, sizey=0.3,
        xanchor="center", yanchor="top"
      )]
    
    sym = RIC.split('.')
    
    
    
    
    
    
    
    return pio.write_html(fig, file='{}_stock.html'.format(sym[0]), auto_open=True), fig.write_image("fig_{}_stock.pdf".format(sym[0]))
   
    
    return 
```
